```python
import os
import re

# Folder containing the text files (assuming it's in the same directory as the notebook)
folder_path = 'Untitled Folder 1'

# Name of the output file
output_file = 'extracted_sections_MD&A#4.txt'

# Get a list of all text files in the folder
file_list = [f for f in os.listdir(folder_path) if f.endswith('.txt')]

# Define the regular expression pattern to match either of the sections and stop at a new section starting with "ITEM" or "head"
pattern = re.compile(
    r"(MANAGEMENT'S DISCUSSION AND ANALYSIS OF FINANCIAL CONDITION AND RESULTS OF OPERATIONS\.?|"
    r"MANAGEMENT S DISCUSSION AND ANALYSIS OF FINANCIAL CONDITION AND RESULTS OF OPERATIONS\.?|"
    r"Management s Discussion and Analysis of Financial Condition and Results of Operations\.?|"
    r"Management's Discussion and Analysis of Financial Condition and Results of Operations\.?|"
    r"Management s Discussion and Analysis of Financial Condition and Results)"
    r"(.*?)(?=^\s*(ITEM|Item|item)|\Z)", 
    re.DOTALL | re.MULTILINE
)

# Open the output file in write mode
with open(output_file, 'w') as outfile:
    for fname in file_list:
        file_path = os.path.join(folder_path, fname)
        # Open each file in read mode
        with open(file_path, 'r') as infile:
            content = infile.read()
            # Search for the section using the regular expression
            matches = pattern.findall(content)
            for match in matches:
                # match[0] is the header, match[1] is the body
                header, body = match[0], match[1]
                # Write each match to the output file
                outfile.write(header + body.strip() + '\n\n')  # Add extra newlines between sections
        # Add a separator line after each file's content
        outfile.write('-----------------------------------------------------------------------------------------------\n\n')
        

print(f'Sections have been extracted and combined into {output_file}')

```

```python
import csv

file_name = '15000firms.csv'
output_file = '15000firms_updated.csv'  # Name of the output file

# Open the CSV file for reading
with open(file_name, 'r') as infile:
    reader = csv.reader(infile)
    
    # Read the header row
    headers = next(reader)
    cik_index = headers.index("cik")
    
    # Open the output file for writing
    with open(output_file, 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        
        # Write the header to the output file
        writer.writerow(headers)
        
        # Iterate through the rows
        for row in reader:
            # Add "0000" in front of the cik value
            row[cik_index] = '0000' + row[cik_index]
            # Write the modified row to the output file
            writer.writerow(row)

print(f'Updated file has been saved as {output_file}')
```

```python
import csv
import pandas as pd

file_name = '15000firms_updated.csv'

# Initialize lists to store the cik and conm values
list_cik_new = []
list_comp_name = []

# Open the CSV file
with open(file_name, 'r') as file:
    reader = csv.reader(file)
    
    # Read the header row to find the index of the "cik" and "conm" columns
    headers = next(reader)
    cik_index = headers.index("cik")
    conm_index = headers.index("conm")
    
    # Iterate through the rows and store the data under the "cik" and "conm" columns
    for row in reader:
        list_cik_new.append( row[cik_index])
        list_comp_name.append(row[conm_index])

# Create a DataFrame from the lists
df = pd.DataFrame({
    'cik': list_cik_new,
    'conm': list_comp_name
})

# Print the DataFrame
print(df)

# Optionally, save the DataFrame to a new CSV file
df.to_csv('updated_firms.csv', index=False)
```
